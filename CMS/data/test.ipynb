```python
import numpy as np
import os

def load_and_print_npz(file_path):
    # 加载 .npz 文件
    data = np.load(file_path, allow_pickle=True)
    
    print("NPZ file contents:")
    print("------------------")
    
    # 遍历并打印文件中的每个数组
    for key in data.files:
        array = data[key]
        print(f"\nArray name: {key}")
        print(f"Shape: {array.shape}")
        print(f"Data type: {array.dtype}")
        
        # 打印数组的前几个元素（如果是一维数组）或第一行（如果是多维数组）
        if array.ndim == 1:
            print("First few elements:", array[:5])
        else:
            print("First row:", array[0])
        
        # 如果是字符串数组，打印一些示例
        if array.dtype.kind in ['U', 'S']:  # U for Unicode, S for byte string
            print("Sample strings:")
            for s in array[:5]:
                print(f"  {s}")

    print("\nEnd of NPZ file contents")

# 使用示例
laion_beyond_root = '/202331510162/llm/hx/LAION_Beyond/'  # 替换为实际路径
args_subfield = "Pokemon"  # 替换为实际的 subfield 值
npz_path = os.path.join(laion_beyond_root, f'{args_subfield}_filtered_clustering_results.npz')

load_and_print_npz(npz_path)
```
